Import Libraries to assist with Linear Regression.

```python
import numpy as np
from numpy import array
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

Read the Insurance csv file and display the first 5 rows of our dataframe.

```python
our_data = pd.read_csv("Insurance.csv")
our_data.head()
```

Use the iloc function to select age and charges from the dataframe and reshape so it is a one dimensional array

```python
x = our_data.iloc[:,0].values
y = our_data.iloc[:,-1].values

y=y.reshape(-1, 1)
x=x.reshape(-1, 1)

```

Using the scikit-learn package to fit a simple linear regression model we can predict along a line by inputting any x value. We can see how similar those predictions are to the x values we know the corresponding y values for. Below these predictions are shown in red, and the obversed values are in blue

```python
age_model = LinearRegression()
age_model.fit(x,y)

y_pred = age_model.predict(x)

plt.scatter(x,y,color = 'b')
plt.plot(x,age_model.predict(x),color = 'r')
plt.ylabel('Charges')
plt.xlabel('Age')
plt.show()
```

Above we have a scatter plot of the data points. We can observe that the higher in age a person is, the higher their charges for insurance. This is a linear relationship that can be estimated with simple linear relationship.

Now I want to use this age_model to predict how much insurance charges will be if you are aged 60 years old

```python
unk_x = [[60]]

x_pred = np.append(x, unk_x).reshape(-1,1)
y_pred = age_model.predict(x_pred)

plt.scatter(x,y,color = 'b')
plt.plot(x_pred,y_pred,color = 'r')
plt.title ('Age v/s Charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()

print("At age 60 the charges for insurance should be:", age_model.predict(unk_x))
```

The above age_model has predicted that at age 60 the charges for insurance should be £18629.24
